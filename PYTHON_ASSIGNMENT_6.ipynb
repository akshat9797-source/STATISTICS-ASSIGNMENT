Q1 What are Type I and Type II errors in hypothesis testing, and how do they impact decision-making?
Ans Type I vs Type II Errors (Hypothesis Testing)
Type I Error (False Positive):
Rejecting a true null hypothesis (seeing an effect that isn't there).
Type II Error (False Negative):
Failing to reject a false null hypothesis (missing a real effect).
Probability Terms:
Type I → α (alpha), significance level
Type II → β (beta), related to test power
Impact on Decisions:
Type I → Leads to wrong actions based on false alarms
Type II → Leads to missed opportunities or undetected problems
Trade-off:
Reducing one error often increases the other (balance is needed).

Q2 What is the P-value in hypothesis testing, and how should it be interpreted in the context of the null hypothesis?
Ans P-value in Hypothesis Testing Probability of observing results as extreme as (or more extreme than) the data, assuming the null hypothesis (H₀) is true.
Core Idea: Measures how compatible your data is with H₀.
Decision Rule:
If p ≤ α (e.g., 0.05) → Reject H₀
If p > α → Fail to reject H₀
Interpretation:
Small p-value → Data is unlikely under H₀ → evidence against H₀
Large p-value → Data is plausible under H₀ → insufficient evidence to reject
Important Clarification:
It is NOT the probability that H₀ is true
It is about the data given H₀, not H₀ given the data
Practical Insight:
Helps decide whether an observed effect is statistically significant.
Q3 Explain the difference between a Z-test and a T-test, including when to use each.
Ans Z-test vs T-test
Both test hypotheses about population means.
Key Difference (Variance Knowledge):
Z-test: Population variance (σ) is known
T-test: Population variance is unknown (use sample estimate)
Sample Size:
Z-test: Typically large samples (n ≥ 30)
T-test: Works well for small samples
Distribution Used:
Z-test: Standard normal distribution
T-test: Student's t-distribution (accounts for extra uncertainty)
Flexibility:
Z-test: More rigid assumptions
T-test: More practical in real-world data
When to Use:
Use Z-test when σ is known and sample is large
Use T-test when σ is unknown (most common case)

Q4 What is a confidence interval, and how does the margin of error influence its width and interpretation?
Ans Confidence Interval (CI) A range of values used to estimate a population parameter (e.g., mean).
Confidence Level: Indicates reliability (e.g., 95% CI → method captures true value in 95% of samples).
Structure:
CI = estimate ± margin of error
Margin of Error (MoE):
Measures uncertainty around the estimate
Depends on sample size, variability, and confidence level
Impact on Width:
Larger MoE → wider interval (less precise)
Smaller MoE → narrower interval (more precise)
Interpretation:
Wider CI → more uncertainty but higher confidence
Narrower CI → more precision but possibly lower confidence

Q5 Describe the purpose and assumptions of an ANOVA test. How does it extend hypothesis testing to more than two groups?
Ans ANOVA (Analysis of Variance)Purpose:
Tests whether the means of 3 or more groups are significantly different.
Core Idea:
Compares between-group variation vs within-group variation.
Hypotheses:
H₀: All group means are equal
H₁: At least one group mean is different
Extension Beyond Two Groups:
Instead of multiple t-tests (which increase error), ANOVA uses a single test to compare many groups efficiently.
Test Statistic:
Uses the F-ratio (variance between groups / variance within groups).
Key Assumptions:
Data is normally distributed
Groups have equal variances (homogeneity)
Observations are independent
Decision Insight:
If significant, it tells a difference exists, but not which groups differ (post-hoc tests needed).




```python
# Q6 Write a Python program to perform a one-sample Z-test and interpret the
# result for a given dataset.


import numpy as np
from scipy import stats

# Sample data
data = [52, 55, 50, 53, 54, 56, 52, 51, 55, 53]

# Known population parameters
pop_mean = 50
pop_std = 4   # known standard deviation

# Calculate sample statistics
sample_mean = np.mean(data)
n = len(data)

# Compute Z-score
z = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))

# Compute p-value (two-tailed)
p_value = 2 * (1 - stats.norm.cdf(abs(z)))

# Significance level
alpha = 0.05

# Output
print("Z-score:", z)
print("P-value:", p_value)

# Interpretation
if p_value < alpha:
    print("Reject the null hypothesis (significant difference).")
else:
    print("Fail to reject the null hypothesis (no significant difference).")


# Q7 Simulate a dataset from a binomial distribution (n = 10, p = 0.5) using
# NumPy and plot the histogram.

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Parameters
n = 10
p = 0.5
size = 1000   # number of samples

# Generate data
data = np.random.binomial(n, p, size)

# Plot histogram
sns.histplot(data, bins=11, kde=False)

plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")

plt.show()

# Q8  Generate multiple samples from a non-normal distribution and implement
# the Central Limit Theorem using Python.


import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

population = np.random.exponential(scale=2, size=10000)

sample_means = []
sample_size = 30
num_samples = 1000

for _ in range(num_samples):
    sample = np.random.choice(population, size=sample_size)
    sample_means.append(np.mean(sample))

sns.histplot(sample_means, bins=30, kde=True)

plt.title("Central Limit Theorem Demonstration")
plt.xlabel("Sample Means")
plt.ylabel("Frequency")

plt.show()


# Q9 Write a Python function to calculate and visualize the confidence interval
# for a sample mean.

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def plot_confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)

    h = std_err * stats.t.ppf((1 + confidence) / 2, n - 1)

    lower = mean - h
    upper = mean + h

    print(f"Mean: {mean}")
    print(f"{int(confidence*100)}% CI: ({lower}, {upper})")

    plt.figure()
    plt.hist(data, bins=10, alpha=0.6)
    plt.axvline(mean)
    plt.axvline(lower)
    plt.axvline(upper)

    plt.title("Confidence Interval Visualization")
    plt.xlabel("Data")
    plt.ylabel("Frequency")

    plt.show()

sample_data = [10, 12, 13, 12, 11, 14, 13, 12, 11, 15]
plot_confidence_interval(sample_data)


# Q10  Perform a Chi-square goodness-of-fit test using Python to compare
# observed and expected distributions, and explain the outcome.


from scipy.stats import chisquare

observed = [50, 30, 20]
expected = [40, 40, 20]

chi_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print("Chi-square Statistic:", chi_stat)
print("P-value:", p_value)

alpha = 0.05
if p_value < alpha:
print("Reject H0: Observed distribution differs from expected.")
else:
print("Fail to reject H0: No significant difference.")
```
